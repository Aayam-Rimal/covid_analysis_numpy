Analysis of covid19 cases using numpy:

```python
import numpy as np

data=np.genfromtxt(
    "country_wise_latest.csv",
    delimiter=",",
    dtype=None,
    names=True,
    encoding="utf-8"
)


```

Assigning data to variables:

```python
country= data["CountryRegion"]
confirmed= data["Confirmed"]
Deaths= data["Deaths"]
Recovered= data["Recovered"]
Active= data["Active"]
```

Mean of different data:

```python
print("--MEAN--\n")
print(f"average confirmed cases: {confirmed.mean()}")
print(f"Average Deaths: {Deaths.mean()}")
print(f"Average recovered cases: {Recovered.mean()}")
print(f"Average active cases: {Active.mean()}")
```

Median of different data :

```python
print("--MEDIAN--\n")
print(f"median confirmed cases: {np.median(confirmed)}")
print(f"median Deaths: {np.median(Deaths)}")
print(f"median recovered cases: {np.median(Recovered)}")
print(f"median active cases: {np.median(Active)}")
```

Standard Deviation of different data:

```python
print("--STANDARD DEVIATION--\n")
print(f"std of confirmed cases: {confirmed.std()}")
print(f"std of Deaths: {Deaths.std()}")
print(f"std of recovered cases: {Recovered.std()}")
print(f"std of active cases: {Active.std()}")
```

Lowest,highest trends:


```python
idx= np.argmax(Deaths)
max_death= Deaths[idx]
country_max= country[idx]
print(f" max no of deaths {max_death} occured in {country_max}")

idx2= np.argmin(Deaths)
min_death= Deaths[idx2]
country_min= country[idx2]
print(f" min no of deaths {min_death} occured in {country_min}")
```

Countries with most deaths, least deaths

```python
most_deaths= Deaths[Deaths>20000]
countries= country[Deaths>20000]

for death,nation in zip(most_deaths,countries):
    print(f"{nation}:{death}")
```

```python
least_deaths= Deaths[Deaths<5]
least_countries= country[Deaths<5]

for died,state in zip(least_deaths,least_countries):
    print(f"{state}:{died}")
```

Multiple columns analysis:

```python
print("countries with more deaths than recovered:")
print(f"{country[Deaths > Recovered]}\n")

print("countries whose recovery exceeds 1000 and deaths below 20 ")
mask= (Recovered>1000) & (Deaths < 20)
print(f"{country[mask]} ")
```

Correlation between different values:

```python
corr_1= np.corrcoef(confirmed,Deaths)
corr_2= np.corrcoef(confirmed,Recovered)
corr_3= np.corrcoef(Recovered, Active)

print(f"the correlation between confirmed cases and Deaths is:\n {corr_1}")
print(f"the correlation between confirmed cases and recovered cases are:\n {corr_2}")
print(f"the correlation between Recovered cases and Active cases are:\n {corr_3}")
```
